```python
areas = ["GB", "Nordic", "GB-CE", "Nordic-CE"]
targets = {
    "GB": [
        "EWIC",
        "Moyle",
    ],
    "Nordic": [
        "Nordbalt",
        "Estlink",
    ],
    "GB-CE": [
        "IFA",
        "Britned",
    ],
    "Nordic-CE": [
        "Norned",
        "Swepol",
        "Baltic Cable",
        "Storebaelt",
        "Skagerrak",
        "Kontiskan",
        "Kontek",
    ],
}

a = []
links = []
indicators = []
just_indicators = []
forecast = []
full = []

for area in areas:
    for link in targets[area]:
        y_pred = pd.read_hdf(fit_folder.format(area, version, link) + "y_pred_links.h5")
        y_test = pd.read_hdf(
            version_folder.format(area, link, version) + "y_test_links.h5"
        )

        print("full: ", r2_score(y_test, y_pred.loc[:, "gtb_full"]))

        links.append(link)
        full.append(r2_score(y_test, y_pred.loc[:, "gtb_full"]))
        a.append(
            area_name[area]
            .replace("/", "<br>")
            .replace("IE", "Ireland")
            .replace("GB", "Great Britain")
            .replace("CE", "Continental Europe")
        )

fig = go.Figure(
    data=[
        go.Bar(
            # name='full',
            x=[a, links],
            y=full,
            offsetgroup=0,
            marker_color=px.colors.qualitative.T10[2]
            # "rgb(255,0,81)",
        ),
    ],
)

fig.update_layout(template="plotly_white", width=1240, height=750, showlegend=False)
fig.update_xaxes(
    linewidth=1,
    linecolor="black",
    title="",
    mirror=True,
    ticks="outside",
    showline=True,
    gridwidth=0,
    tick0=0,
    dtick=1,
)
fig.update_yaxes(
    linewidth=1,
    linecolor="black",
    title="R<sup>2</sup>",
    mirror=True,
    ticks="outside",
    showline=True,
    gridwidth=1,
    range=[0, 0.95],
    title_standoff=40,
    tickmode="array",
    tickvals=[0, 0.2, 0.4, 0.6, 0.8],
)

fig.update_layout(font=dict(size=18, color="black"))
fig.update_annotations(font_size=20)
fig.update_layout(legend_font_size=20)
fig.update_layout(margin_t=0)
fig.update_layout(margin_r=50)

print("creating plot...")
scope = pio.kaleido.scope
scope._shutdown_kaleido()
pio.write_image(fig, "fig6.pdf", format="pdf", validate=False, engine="kaleido")
time.sleep(1)
pio.write_image(fig, "fig6.pdf", format="pdf", validate=False, engine="kaleido")
scope._shutdown_kaleido()
```
